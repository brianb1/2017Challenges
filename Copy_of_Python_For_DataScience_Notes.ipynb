***<center> Notes for Python for Data Science  (Intro to NumPy and Pandas)</center> ***

```
#Importing Packages
import numpy as np
import pandas as pd 
```

**NumPy Arrays are like Lists but allow for Arthmetic Operations to be performed easily and faster**

```
#Suppose we have the 2 lists below representing revenue and cost data for 8 quarters below.
#The first 4 elements represent the 1st year's quarterly data.
#The last 4 elements represent the 2nd year's quarterly data. 
rev_list = [300,350,450,550,320,340,440,580]
cost_list = [250,280,320,360,280,270,340,400]

#Suppose we want to calculate the profit and the following line of code:
#Profit = rev - cost 
#This will throw an error. NumPy arrays on the otherhand allow us to perform arthimetic operations.

#How to create a numpy array => np.array(Insert a list here)
#Note that unlike a list, all numpy array elements have to be of the same datatype. 

#For example:

rev_array = np.array(rev_list)
cost_array = np.array(cost_list)

#Now we can calculate the profit more easily:
profit_array = rev_array - cost_array 
print(profit_array)

#You can perform standard arithmetic operations on numpy arrays such as +,-,*,/ etc.

#You can filter arrays as well. 
#This prints out only the values where the element is greater than 300. 

print(rev_array[rev_array>300])
```

**<center>Shaping Data Using NumPy</center>**

```
#You can view the shape of your array be assessing the .shape property. 
print(rev_array.shape) #Note (8,) indicates 8 rows and 0 columns

#You need to use the reshape method on your array if we you want to structure it. 

print(rev_array.reshape(2,4)) #This results in a 2-d array with 2 rows and 4 columns

#Note that when you use the reshape method, the underlying array is left unchanged.
#One way you can still assess the product of the reshape method is to assign it to a variable.

rev_array_reshaped = rev_array.reshape(2,4,1) #We are reshaping the array to be 3-d here.
print(rev_array_reshaped)
```

**<center> Slicing Your Data -AKA Getting What You Want from the Data </center>**


```
#The syntax for slicing your data is similar to slicing data from lists. 
#Note that python indexing starts with 0. 

first_quarter_first_year = rev_array[0]

print(first_quarter_first_year)

#Suppose we want to get first three quarters' revenue 

print(rev_array[:3]) #0,1,2 elements gets outputted 
```

**<center> Moving onto Pandas- How do you create a DataFrame? </center> **

```
#In Pandas, we work with dataframes.

#One way of you can get data is to convert dictionary into a dataframe 

sample_dictionary = [{'a':1,'b':2,'c':3,'d':4},
                    {'a':100,'b':200,'c':300,'d':400}]

df_from_dictionary = pd.DataFrame(sample_dictionary)

print(df_from_dictionary)
```

```
#You can also import data from csv, html, xlsx (excel) formats to create a dataframe.

#Suppose you want to read from a csv file. You use Panda's read_csv method. 

df = pd.read_csv("https://raw.githubusercontent.com/gpalrepo/inclass_practice/master/Shark%20Tank%20Companies.csv")
```

**<center>Some Exploratory Data Analysis</center>**

```
print(df.shape) #shape property indicates number of rows and columns in your dataframe.
print(df.info()) # Info method tells you the columns and associated datatypes. 
```

```
print(df.head()) #This prints out the first 5 rows in a dataset
print(df.tail()) #This prints out the last 5 rows in a dataset
```

```
print(df.columns) #This prints out just the names of the columns 

#You can rename a column's name using the rename method. 
#You need to make sure the key is the orignal name of the column 
#and the value is the new name. 

#Inplace=True ensures that you make the changes to the existing dataframe itself.

df.rename(columns={'website':'company_website'},inplace=True) 

#Let's check to see the column names have been changed.
print(df.columns)
```

**<center> Slicing DataFrames Using iloc and loc </center> **

```
#How we can access specific rows within our dataframe? 
#iloc method => integer based location indexing

print(df.iloc[0,1]) #Prints out the 1st row and 2nd column. 

print(df.iloc[1:4,5:9]) #Prints out 2nd to 4th row and 6th to 9th row.  
```

```
#iloc isn't intuitive. At times, it better to use label-based filtering.
#The loc method allows us to do that 

#Suppose we want to print the 'valuation' and 'title' columsn of
#all the rows which belong to the Electronics category 

print(df.loc[df['category']=='Electronics',['valuation','title']])

#We can also filter results by boolean values. This outputs all rows in which
#the valuation is greater than 1000000.

print(df.loc[df['valuation']>1000000])
```

**<center>Performing Mathematical Operations using Pandas </center> **

```
#Sample Mathematical Operations 
#We are performing mathematical operations on the 'valuation' column for the examples.
print(df['valuation'].mean())
print(df['valuation'].median())
print(df['valuation'].std())
print(df['valuation'].max())
print(df['valuation'].min())
print(df['valuation'].count())

#Group By 

#When you apply the groupby function, you are 
#i) Splitting the data into groups based on the criteria you specify
#ii) Applying a function to each group
#iii) Finally, combined the results into one output


#The code below helps us find the median valuation across categories. 

print(df.groupby(['category']).median()['valuation'])

```
